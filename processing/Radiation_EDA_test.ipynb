```python
# 06시 ~ 18시
mask = origin_df['Time'].isin(['23', '22', '21', '20', '19', '05', '04', '03' ,'02', '01', '00'])

# 마스크 영역을 제외한 영역
df = origin_df[~mask]
df.head()
```

```python
df_small = df.drop(labels=['UNIXTime', 'Data', 'TimeSunRise', 'TimeSunSet'],axis=1)
df_small = df_small.astype({'Time':'int'})
df_small.head()
```

```python
colormap = plt.cm.PuBu
plt.figure(figsize=(10, 10))
plt.title("Person Correlation of Features", y = 1, size = 20)
sns.heatmap(round(df_small.astype(float).corr(), 3), linewidths = 0.1, vmax = 1.0,
           square = True, cmap = colormap, linecolor = "white", annot = True, annot_kws = {"size" : 10})
```
